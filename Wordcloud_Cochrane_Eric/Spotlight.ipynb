# Dealing with Text Data

I began this project full of hope and great expectations to create a flawless classifier with nltk because adding a bit of Natural Language Processing, and basic English comprehension should optimize classifier, right?

I could not have been more wrong and NLTK is not the golden ticket to solving all of your text processing problems...

```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
import pandas as pd
from HTMLParser import HTMLParser
import re
import nltk
from nltk.corpus import stopwords
```

```python
class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()

def clean(text):
    removed_html=strip_tags(text)
    letters_only = re.sub("[^a-zA-Z]", " ", removed_html) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops] 
    return(meaningful_words)

df = pd.read_csv('Questions.csv',encoding = 'iso-8859-1')
df = df.iloc[0:100,]
```

```python
print df.shape
#test to see if removing html worked.
df['Non_html_body']=df['Body'].apply(lambda x:strip_tags(x.encode('utf-8')))
#Remove stopwords, lower all chars, removed html
df['words'] = df['Body'].apply(lambda x:clean(x.encode('utf-8')))
print df['words']
```

# Visualizing Text data with Wordcloud
There is a super handy library for visualizing text data as wordclouds, and yep you guessed it, it runs on matplotlib.

http://peekaboo-vision.blogspot.de/2012/11/a-wordcloud-in-python.html
Repo:
https://github.com/amueller/word_cloud

Another really cool feature about this wordcloud aside from the ease of use is the feature to mask our word clouds with a background image.

```python
import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import random

%matplotlib inline

#wordcloud has its own list of stopwords but I have a hunch they are from nltk
from wordcloud import WordCloud, STOPWORDS

#override color function for the storm trooper mask.
# the default color scheme is rainbow.
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

#d = path.dirname(__file__)

# read the mask image
mask = np.array(Image.open(path.join('', "stormtrooper_mask.png")))

# movie script of "a new hope"
text = open("a_new_hope.txt").read()

# preprocessing the text a little bit
text = text.replace("HAN", "Han")
text = text.replace("LUKE'S", "Luke")

# adding stopwords
stopwords = set(STOPWORDS)
#these appear in the movie script a lot so they should be removed
stopwords.add("int")
stopwords.add("ext")


wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10,
               random_state=1).generate(text)
# store default colored image
default_colors = wc.to_array()
plt.title("Custom colors")
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3))
wc.to_file("a_new_hope.png")
plt.axis("off")
plt.figure()
plt.title("Default colors")
plt.imshow(default_colors)
plt.axis("off")
plt.show()
```
