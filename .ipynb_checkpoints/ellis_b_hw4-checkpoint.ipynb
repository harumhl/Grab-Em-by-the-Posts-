## (70 points) Part 1: Sentiment Analysis

Our goal for this first part is to perform Sentiment Analysis: classifying an entire movie review as positive or negative. In particular, we are going to analyze the text of 1000 positive movie reviews and 1000 negative reviews, toward building an automatic sentiment polarity analyzer. **Task 1:** Your first task is just to load the reviews. You should consider the first 500 positive and the first 500 negative reviews as your training set. You should consider the remaining reviews as your test set.

```python
# Task 1: Load the movie reviews
#source - http://stackoverflow.com/questions/3964681/find-all-files-in-directory-with-extension-txt-in-python
import glob, os
import pandas as pd

#read the files
pos = []
os.chdir("/Users/Brandon_Ellis/Desktop/Fall 16/cs 489/HW4/review_polarity/txt_sentoken/pos/")
for file in glob.glob("*.txt"):
    f = open(file, 'r')
    pos.append(f.read())

neg = []
os.chdir("/Users/Brandon_Ellis/Desktop/Fall 16/cs 489/HW4/review_polarity/txt_sentoken/neg/")
for file in glob.glob("*.txt"):
    f = open(file, 'r')
    neg.append(f.read())
    
pos_lable = []
neg_lable = []
for i in range(0,1000):
    pos_lable.append("pos")
    neg_lable.append("neg")

# set up the training data with first 500 files
pos_train = pd.DataFrame({'content':pos[0:500],'target':pos_lable[0:500]})
neg_train = pd.DataFrame({'content':neg[0:500],'target':neg_lable[0:500]})
train = pos_train.append(neg_train)
train.reset_index(inplace=True)
train = train[['content','target']]
#print train.shape
    
# set up the testing data with last 500 files
pos_test = pd.DataFrame({'content':pos[500:1000],'target':pos_lable[500:1000]})
neg_test = pd.DataFrame({'content':neg[500:1000],'target':neg_lable[500:1000]})
test = pos_test.append(neg_test)
test.reset_index(inplace=True)
test = test[['content','target']]
#print test.shape
```

**Task 2:** Now you should represent each review as a vector, where element i of the vector corresponds to the count of word i in that movie review. This is the classic `bag-of-words` model. It turns out scikit-learn comes built-in with a nice helper called *CountVectorizer* that does exactly what you need. Something like:

`from sklearn.feature_extraction.text import CountVectorizer`

`tf_vectorizer = CountVectorizer()`

`tf = tf_vectorizer.fit_transform(dataset)`

For this part, you should call CountVectorizer with its default settings. 

```python
# Task 2: 
# Use CountVectorizer to parse the reviews
from sklearn.feature_extraction.text import CountVectorizer
tf_vectorizer = CountVectorizer()

#tf_pos = tf_vectorizer.fit_transform(pos_train)
#tf_neg = tf_vectorizer.fit_transform(neg_train)
#tf_pos = tf_vectorizer.fit_transform(pos)

#dataset = pos + neg
#dataset = pos_train + neg_train

tf_train = tf_vectorizer.fit_transform(train['content'].values)
tf_test = tf_vectorizer.transform(test['content'].values)
```

**Task 3:** Now we're going to learn a classifier over the training set. In this case, use `sklearn.naive_bayes.MultinomialNB`. Multinomial Naive Bayes is a popular classifier for use in text classification. Then apply this learned classifier over your test set. Report the accuracy of this approach (using `.score`)

```python
# Task 3: 
# Build a MultinomialNB classifier for determining if 
# a review is positive or negative and evaluate it 
# with accuracy over a 50/50 train-test split.

from sklearn.naive_bayes import MultinomialNB
# piazza source
clf = MultinomialNB().fit(tf_train, train['target'])
print clf.score(tf_test, test['target'])
```

## (30 points) Part 2: Clustering

For this second part, we're going to get our hands dirty with clustering over a collection of baseball statistics. You should begin by grabbing this amazing collection of baseball statistics:

http://seanlahman.com/files/database/baseballdatabank-master_2016-03-02.zip

```python
# first, let's load the data for players 
%matplotlib inline 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# you may need to update the data path in read_csv
# to reflect wherever you store your data
all_players = pd.read_csv('/Users/Brandon_Ellis/Desktop/Fall 16/cs 489/HW4/baseballdatabank-master/core/Batting.csv') 
fielding = pd.read_csv('/Users/Brandon_Ellis/Desktop/Fall 16/cs 489/HW4/baseballdatabank-master/core/Fielding.csv')

all_players.shape
```

```python
# You should see that there are 101,332 rows with 22 columns 

# To focus our study, we're going to keep players who played
# since 1990 and had at least 500 plate appearances
players = all_players[(all_players.AB + all_players.BB > 500)  & (all_players.yearID > 1990)].copy()
players.shape
```

```python
# we also need to get the player position into our players dataframe
# this should add one extra column with the position
from collections import defaultdict

def find_pos(df):
    positions = df.POS
    d = defaultdict(int)
    for pos in positions:
        d[pos] += 1
    result = max(d.iteritems(), key=lambda x: x[1])
    return result[0]

positions_df = fielding.groupby("playerID").apply(find_pos)
positions_df = positions_df.reset_index()
positions_df = positions_df.rename(columns={0:"POS"})
players = positions_df.merge(players, how='inner', on="playerID")
players.shape
```

```python
list(players.columns.values)
```

```python
players.head()
```

```python
# Okay, now we're down to 3,181 rows (corresponding to a player in a particular year)

# You'll notice that there's no column for singles (1B), so we
# can calculate it based on the total hits - doubles - triples - homers
players["1B"] = players.H - players["2B"] - players["3B"] - players["HR"]

# IGNORE FOR NOW
# We can also create a new column for plate appearances = walks + at-bats
players["PA"] = players.BB + players.AB

# Then we can define the rate of each kind of hit or walk per plate appearance
for col in ["1B","2B","3B","HR","BB"]:
    players[col+"_rate"] = players[col]/players.PA

# We can create a new dataframe with these stats + their position
playerstats = players[["playerID","yearID","1B","2B","3B","HR","BB","POS","PA","1B_rate","2B_rate","3B_rate","HR_rate","BB_rate","H","AB","BB","IBB", "HBP", "SF","lgID","RBI"]].copy()
playerstats.head()
```

```python
# We can map each position to a number
# for simplicity, we'll map some positions to the same number
positions = {'P' : 1,
             'C' : 2,
             '1B' : 3,

             '2B' : 4,
             'SS' : 4,
             
             '3B' : 5,

             'LF' : 7,
             'CF' : 7,
             'RF' : 7,
             'OF' : 7,
             'DH' : 8}

# and then update the dataframe to include a new column
playerstats['POS_Numeric'] = playerstats.POS.map(positions)
playerstats.head()
```

```python
# Now, let's do some clustering!
from sklearn.cluster import KMeans
cluster_alg = KMeans(n_clusters=10)

# what featurs should we use to represent each player?
X = playerstats[['1B', 'HR']]

# do the clustering
cluster_alg.fit(X)

# these are the cluster labels
labels = cluster_alg.labels_

plt.figure(figsize=(20,10))
plt.scatter(X['1B'], X['HR'], c=labels.astype(np.float), cmap=plt.cm.Paired, s=200)
plt.xlabel('1B')
plt.ylabel('HR')
```

```python
# we can see that there are nice clusters discovered. 
# but do they actually correspond to the positions the players
# play? let's take a look at the same data but with colors
# based on positions:

labels = playerstats['POS_Numeric'].reshape(-1)

plt.figure(figsize=(20,10))
plt.scatter(X['1B'], X['HR'], c=labels.astype(np.float), cmap=plt.cm.Paired, s=200)
plt.xlabel('1B')
plt.ylabel('HR')
```

```python
# uh oh, we can see that the positions don't really correspond
# to the clusters based on homeruns + singles

# we can quantify this error using a clustering metric
# like AdjustedRandIndex (which ranges from -1 to 1, with 1 being perfect, that is
# each cluster is composed only of elements from the same underlying
# class -- like position in this case)
from sklearn import metrics
print metrics.adjusted_rand_score(labels, cluster_alg.labels_)
```

```python
# some more "calculations" adding more columns to df
# source - http://www.espn.com/mlb/stats/batting/_/league/al
# to find what other types of statistics we could potnetially use

#total bases touched
playerstats['total_bases'] = players['1B']+2*players['2B']+3*players['3B']+4*players['HR']
# Batting Average (BA) = ( Number of hits ) / ( Number of official at bats ) 
playerstats['BA'] = playerstats['H'] / playerstats['AB']
# Slugging (SLG)
playerstats['SLG'] = playerstats['total_bases'] / playerstats['AB']
# On base Percentage (OBP) = (Hits + Walks + Hit by Pitch) / (At Bats + Walks + Hit by Pitch + Sacrifice Flies)
# source - http://www.homerunweb.com/onbase.html
playerstats['walks'] = playerstats['IBB'] + playerstats['HBP']
playerstats['OBP'] = (playerstats['H'] + playerstats['walks']) / (playerstats['AB'] + playerstats['walks'] + playerstats['SF'])
#on-base plus slugging (OPS)
# source - https://www.easycalculation.com/sports/baseball-ops-calculator.php
playerstats['OPS'] = playerstats['SLG'] + playerstats['OBP']

#convert league ID to numeric
league_to_num = {'NL':1, 'AL':2}
playerstats['lg_Numeric'] = playerstats.lgID.map(league_to_num)

playerstats
```

```python
# what featurs should we use to represent each player?
xFeat = 'RBI'
yFeat = 'BA'

X = playerstats[[xFeat, yFeat]]

#cluster_alg = KMeans(n_clusters=10, max_iter=300, n_init=10)
cluster_alg = KMeans(n_clusters=5, max_iter=300, n_init=10)
cluster_alg.fit(X)
labels = cluster_alg.labels_

```

```python
plt.figure(figsize=(20,10))
plt.scatter(X[xFeat], X[yFeat], c=labels.astype(np.float), cmap=plt.cm.Paired, s=200)
plt.xlabel('HR')
plt.ylabel('OPS')
print metrics.adjusted_rand_score(labels, cluster_alg.labels_)
```

```python
labels = playerstats['POS_Numeric'].reshape(-1)

plt.figure(figsize=(20,10))
plt.scatter(X[xFeat], X[yFeat], c=labels.astype(np.float), cmap=plt.cm.Paired, s=200)
plt.xlabel('HR')
plt.ylabel('OPS')
```

```python
print metrics.adjusted_rand_score(labels, cluster_alg.labels_)
```
